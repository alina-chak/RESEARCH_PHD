```python
from pyswarm import pso

def banana(x):
    x1 = x[0]
    x2 = x[1]
    return x1**4 - 2*x2*x1**2 + x2**2 + x1**2 - 2*x1 + 5

def con(x):
    x1 = x[0]
    x2 = x[1]
    return [-(x1 + 0.25)**2 + 0.75*x2]

lb = [0, 0.5]
ub = [0.5, 1]

xopt, fopt = pso(banana, lb, ub, f_ieqcons=con)

print(xopt)
print(fopt)

#objective function will be the precision for n number of videos retrieved
```
